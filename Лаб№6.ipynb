```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_01.
#
# Выполнила: Вереина М.С.
# Группа: АДЭУ-211
# E-mail: VereinaMS@mgpu.ru

"""Алгоритм выполняет сортировку списка по возрастанию

Сложность: O(N) * O(N) * (O(1) + O(1)) = O(N^2)"""

def foo(a):
    for i in range(len(a), 0, -1):    #O(N)
        for j in range(1, i):         #O(N)
            if a[j-1] > a[j]:         #O(1)
                a[j-1], a[j] = a[j], a[j-1]   #O(1)
    return a


a = [1, 2, 3, 4, 2, 1, 3, 4, 3, 5, -2, -2, 2, -2, 65, 2, 2]
foo(a)
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_02.
#
# Выполнила: Вереина М.С.
# Группа: АДЭУ-211
# E-mail: VereinaMS@mgpu.ru

"""Алгоритм должен вывести то же самое число

Сложность: O(N)+O(1)=O(N+1)=O(N) """

def foo(i):
    digits = "0123456789"
    if i == 0:                #O(1)
        return "0"
    result = ""
    while i > 0:              #O(N)
        result = digits[i%10] + result
        i = i // 10
    return result

i = 67
foo(i)
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_03.
#
# Выполнила: Вереина М.С.
# Группа: АДЭУ-211
# E-mail: VereinaMS@mgpu.ru

"""
    Алгоритм должен вывести сумму чисел, записанных в виде строки

    Сложность: O(N) * O(1) = O(N).
"""
def foo(s):
   
    val = 0
    for c in s:             #O(N)
        if c.isdigit():     #O(1)
            val += int(c)
    return val

s = "5, 6, 7, 8"
foo(s)
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_04.
#
# Выполнила: Вереина М.С.
# Группа: АДЭУ-211
# E-mail: VereinaMS@mgpu.ru

"""Алгоритм должен вывести все простые числа до введенного числа, включая само число, если оно простое

    Сложность: O(N) * (O(1) * O(1) + O(1)) = 2*O(N) = O(N)
"""
  
def foo(n):
    res = []
    for i in range(1, n + 1):       #O(N)
        divisors = 0
        j = 2
        while j < i and divisors == 0:      #O(1)
            if i % j == 0:                  #O(1)
                divisors += 1
            j += 1

        if divisors == 0:                   #O(1)
            res.append(i)

    return res

n = 54
foo(n)
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_06_02_05.
#
# Выполнила: Вереина М.С.
# Группа: АДЭУ-211
# E-mail: VereinaMS@mgpu.ru

"""Алгоритм проверяет четность/нечетность элементов списка и выводит True, если хотя бы один из них четный
   
   Сложность: O(1) + O(1) = 2*O(1) = O(1).
"""
def foo(nums):
    
    for x in nums:        #O(1)
        if x % 2 == 0:    #O(1)
            return True
    else:
        return False

nums = 1, 3, 5, 7, 10
foo(nums)
```
