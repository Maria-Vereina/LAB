```python
# Практическая работа №4. Индивидуальное задание
# Вариант 4
# Вереина Мария, АДЭУ-211
# Механизм - Изделие - Узел/Деталь

class Mechanism:
    def __init__(self, name, guarantee, maker, term):
        self.name = name                #тип механизма
        self.maker = maker              #производитель
        self.term = term                #срок службы
        self.guarantee = guarantee      #срок гарантии
        print ('Гарантии всего: {}, производитель: {}, срок службы: {}'.format(self.guarantee, self.maker, self.term))  

    def mech(self):
        return "Механизм {}".format(self.name)

    def left(self):                     #считаю, сколько осталось гарантии
        L = self.guarantee - self.term
        print('Гарантии осталось {} года(лет)'.format(L))
    
    def action(self):                   #думаю, надо ли обновлять механизм
        if self.guarantee - self.term > 5:
          print("Можно жить спокойно")
        else:
          print("Надо срочно обновлять механизм")

class Product(Mechanism):
    def __init__(self, name, guarantee, maker, term, power):     #в подклассе добавляю новую переменную - мощность
        Mechanism.__init__(self, name, guarantee, maker, term) 
        self.power = power              #мощность

    def prod(self):
        return "Фонарик {}".format(self.name)

    def action(self):                   #полиморфизм(подсчет максимальной мощности)
        M = self.power*250
        print('Максимальная мощность изделия {} Вт'.format(M))

    def check(self):                    #оцениваю производителя
        if self.power*250 > 500000:
          print(f'{self.maker} - хороший производитель')
        else:
          print(f'{self.maker} не дотягивает')

class Node(Product):
    def __init__(self, name, guarantee, maker, term, power, I):     #в подклассе добавляю новую переменную - проводимость тока
        Product.__init__(self, name, guarantee, maker, term, power)
        self.I = I                    #проводимость тока
    
    def node(self):
        return "Припаивается {}".format(self.name)

    def electro(self):                #подсчет электропроводности
        E = self.power/self.I
        print('Электропроводность узла {} Ом'.format(E))

    def life_or_death(self):          #вопрос жизни и смерти
        if self.power/self.I > 300:
          print("Вам конец")
        elif self.power/self.I == 100:
          print("Будьте осторожны")
        else:
          print("Все супер")

class Detail(Product):
    
    def det(self):
        return "Материал - {}".format(self.name)
        
    def check(self):                      #полиморфизм(узнаю, будет ли служить деталь в этом году)
        while self.term < self.guarantee:
          print("Деталь еще будет служить")
          self.term = self.term + 1
          break
        else: 
          print('Деталь пора менять')

    

    def strength(self):                   #считаю прочность детали
      S = self.power/self.term
      print('Прочность детали: {}'.format(S))
```

```python
mech1 = Mechanism("включен", 25, 'Япония', 23)
print(mech1.mech())
mech1.left()
mech1.action()

prod1 = Product("светит", 50, 'Китай', 35, 4000)
print(prod1.prod())
prod1.action()
prod1.check()

node1 = Node("анод", 40, "Италия", 32, 8345, 20)
print(node1.node())
node1.electro()
node1.life_or_death()

det1 = Detail("сталь", 20, 'Россия', 15, 39287)
print(det1.det())
det1.check()
det1.strength()
```
